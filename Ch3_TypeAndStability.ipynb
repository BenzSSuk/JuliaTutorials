# Using type
you can indicate variables or function parameters to be restricted to specific
types using the ::


```julia
print("1, type:$(typeof(1))\n")
print("'benz', type:$(typeof("benz"))\n")
```

```julia
f(x::Integer)=print("x=$x Type:$(typeof(x))")
```

### Function work well when type is correct

```julia
f(1)
```

### Function raise an error when type is not correct

```julia
f(1.2)
```

# Julia's type hierarchy

### Any for all type of variable

```julia
f2(x::Any)=print("x=$x Type:$(typeof(x))")
```

```julia
f2(1)
```

```julia
f2(1.5)
```

### Use AbstractFloat for both of Float32 and Float64

```julia
f_float(x::Float)=print("x=$x Type:$(typeof(x))")
```

```julia
f_Absfloat(x::AbstractFloat)=print("x=$x Type:$(typeof(x))")
```

```julia
f_Absfloat(1.3)
```

```julia
x=Float32(1.6)
f_Absfloat(x)
```

# Struct 

### Immutable struct
struct that can not change values after intialized

```julia
struct personInfo
    gender::String
    age::Int32
end
```

```julia
benz=personInfo("Male",28)
```

```julia
benz.gender
```

```julia
benz.age
```

```julia
benz.age=30
```

### Mutable struct
Editable struct

```julia
mutable struct personInfo2
    gender::String
    age::Int32
end
```

```julia
benz2=personInfo2("Male",28)
```

```julia
benz2.age
```

```julia
benz2.age=29
```
